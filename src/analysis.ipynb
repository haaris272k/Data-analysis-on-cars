```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('US cars.csv')
df
```

```python
# data cleaning
print(df.isnull()) # checking for null values(ex NaN).
df.isnull().sum() # finding sum of null values in a row.
```

```python
df.dropna() #removing rows with any null values

# we can remove a number of columns, if they are of no use to us.
del df['Unnamed: 0']
del df['vin']
del df['lot']
del df['condition']
df
```

```python
# Q-5 Analyse which brand of car was manufactured the most ?

for i in df['brand'].unique():
   print(i,":",len(df[df['brand']  == i]))

# we took the unique brands then found out no of cars in each brand
# we got to know that ford was the leading brand between 1973 and 2020
```

```python
# Q-6 Analyse which model was manufactured the most ?

for j in df['model'].unique():
    print(j,":",len(df[df['model']==j]))
# we can see that door and f-150 type of mdoel were manufactured the most
```

```python
# Q-7  Analyse which type of color of car was manufactured the most ?

for k in df['color'].unique():
    print(k,":",len(df[df['color']==k]))
# Ans- we can see white and black were manu. the most
```

```python
# Main questions begins from here!
# Q-8 find the top 5 brands in terms of number of cars ?
# logic = iterating> conv into list> conv into dataframe> sorting will be used> visualising

lst=[]

for i in df['brand'].unique():                    # iterating through the values in that coloumn.
    lst.append([i , len(df[df['brand']  == i])])  #converting the data into list.
    
conv_dataframe=pd.DataFrame(lst,columns=['name','freq']) # converting the list to dataframe for ease.

top5_brands=conv_dataframe.sort_values(by='freq',ascending=False).head() 
# sorting and getting top 5 brands and storing that dataframe in a new variable.
top5_brands
```

```python
# visualizing the above data using bar graph.
top5_brands.plot(x='name',y='freq',kind='bar',figsize=(8,8))
plt.show()
```

```python
# Q-9 Find the most popular car colors (top 5)
lst2=[]
for n in df['color'].unique():
    lst2.append([n , len(df[df['color']  == n])])
conv_dataframe2=pd.DataFrame(lst2,columns=['color','unit'])
top5_colors=conv_dataframe2.sort_values(by='unit',ascending=False).head()
top5_colors
```

```python
# visualizing the above data using bar graph
top5_colors.plot(x='color',y='unit',kind='bar',figsize=(8,8))
plt.show()
```

```python
# Q-10 Which car brand is having maximum average price.
# groupby() function will come into use.
# It is used for grouping the data according to the categories and apply a function to the categories. 
# It also helps to aggregate data efficiently.

for_mxavg=df.groupby('brand').mean()['price'].sort_values(ascending=False).plot(kind='bar',figsize=(16,6)) #find the max avg price of each brand and sort.
plt.show()
```

```python
# Q-11 Which state is having average of most expensive cars.
df.groupby('state').mean()['price'].sort_values(ascending=False).plot(kind='bar',figsize=(20,6))
plt.show()

# Now we could easily make out the states where the cars should be manufactured to have a maximum profit
```
